````python
import pandas as pd
````

````python
df = pd.read_csv('data/deertrail.csv', parse_dates=['datetime'])
df.head(10)
````

````python
df.dtypes
````

````python
df.shape
df.value.isna().sum()
df.value.isnull().sum()
````

````python
df.loc[df.parameter == 'Gross Beta', ['value']] = 100
df.loc[df.parameter == 'Gross Beta']
````

````python
df['nd'] = df['value'].isna()
df['adj_value'] = df['value'].fillna(value=df['detection_limit']/2)
df.head()
````

````python
dfmi = df.set_index(['location', 'parameter', 'datetime']).sort_index()
dfmi.head(25)
````

````python
dfmi.loc[('BP-1', 'pH'), ['value']
               ].reset_index().plot(x='datetime', y='value', label="pH")
````
